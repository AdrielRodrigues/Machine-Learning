```python
import pandas as pd
import numpy as np
```

```python
from google.colab import drive
drive.mount("/content/drive")
train = pd.read_csv("/content/drive/MyDrive/Kaggle/Titanic/train.csv")
test = pd.read_csv("/content/drive/MyDrive/Kaggle/Titanic/test.csv")
```

```python
train.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
modelo = RandomForestClassifier(n_estimators=100, n_jobs=-1, random_state=0)
```

```python
variaveis = ['Sex_binario', 'Age']
```

```python
def transformar_sexo(valor):
  if valor == 'female':
    return 1
  else:
    return 0

train['Sex_binario'] = train['Sex'].map(transformar_sexo)
```

```python
X = train[variaveis]
Y = train['Survived']
```

```python
X.head()
```

```python
Y.head()
```

```python
X = X.fillna(-1)
```

```python
modelo.fit(X, Y)
```

```python
test['Sex_binario'] = test['Sex'].map(transformar_sexo)
X_prev = test[variaveis]
X_prev = X_prev.fillna(-1)
```

```python
X_prev.head()
```

```python
p = modelo.predict(X_prev)
p
```

```python
sub = pd.Series(p, index=test['PassengerId'], name='Survived')
sub.to_csv("/content/drive/MyDrive/Kaggle/Titanic/first_submission.csv", header=True)
```
